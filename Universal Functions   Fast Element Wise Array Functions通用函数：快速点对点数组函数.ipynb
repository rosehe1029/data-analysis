universal function, 或 ufunc, 是用来在ndarray中实现element-wise操作的。

可以认为这个ufunc可以把一些简单的函数做快速的向量化封装，输入是一个以上的标量，输出也是一个以上的标量。

很多ufuncs都是点对点的变换，像sqrt或exp：

```python
import numpy as np
arr=np.arange(10)
```

```python
arr
```

```python
np.sqrt(arr)
```

```python
np.exp(arr)
```

```python
#这些函数叫做一元通用函数unary ufuncs，其他一些函数，比如add或maximum
```

```python
x=np.random.randn(8)
```

```python
y=np.random.randn(8)
```

```python
y
```

```python
np.maximum(x,y)
```

尽管不常见，但ufunc也能返回多个数组。例如modf，这是一个向量版的divmod（python内建函数），modf会返回小数部分和整数部分：

    本函数是实现a除以b，然后返回商与余数的元组。如果两个参数a,b都是整数，那么会采用整数除法，结果相当于（a//b, a % b)。如果a或b是浮点数，相当于（math.floor(a/b), a%b)。



```python
arr=np.random.randn(7)*5
arr
```

```python
remainder,whole_part=np.modf(arr)
```

```python
remainder
```

```python
whole_part
```

```python
arr
```

```python
np.sqrt(arr)
```

```python
np.sqrt(arr,arr)
```
